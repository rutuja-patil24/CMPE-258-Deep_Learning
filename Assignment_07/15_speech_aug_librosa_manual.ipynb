```python
# Step 2: Imports
import numpy as np
import librosa
import librosa.display
import soundfile as sf
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# Step 3: Load sample audio (using trumpet as speech-like example)
file_path = librosa.example('trumpet')  # built-in audio
y, sr = librosa.load(file_path, sr=None)
```

```python
# Step 4: Define manual augmentations

def add_noise(audio, noise_level=0.005):
    noise = np.random.randn(len(audio))
    return audio + noise_level * noise

def pitch_shift(audio, sr, n_steps=4):
    return librosa.effects.pitch_shift(audio, sr=sr, n_steps=n_steps)

def time_stretch(audio, rate=1.25):
    return librosa.effects.time_stretch(audio, rate=rate)
```

```python
# Step 5: Apply augmentations
noisy_audio = add_noise(y)
pitched_audio = pitch_shift(y, sr, n_steps=4)
stretched_audio = time_stretch(y, rate=1.2)
```

```python
# Step 6: Save audios
sf.write("original.wav", y, sr)
sf.write("noisy.wav", noisy_audio, sr)
sf.write("pitched.wav", pitched_audio, sr)
sf.write("stretched.wav", stretched_audio, sr)
```

```python
# Step 7: Plot waveforms
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
librosa.display.waveshow(y, sr=sr)
plt.title("Original")

plt.subplot(2, 2, 2)
librosa.display.waveshow(noisy_audio, sr=sr)
plt.title("Noisy")

plt.subplot(2, 2, 3)
librosa.display.waveshow(pitched_audio, sr=sr)
plt.title("Pitched +4 semitones")

plt.subplot(2, 2, 4)
librosa.display.waveshow(stretched_audio, sr=sr)
plt.title("Time Stretched (1.2x)")

plt.tight_layout()
plt.show()
```

```python
# Step 8: Listen to them
print(" Original:")
ipd.display(ipd.Audio("original.wav"))

print(" Noisy:")
ipd.display(ipd.Audio("noisy.wav"))

print(" Pitched:")
ipd.display(ipd.Audio("pitched.wav"))

print(" Stretched:")
ipd.display(ipd.Audio("stretched.wav"))
```
