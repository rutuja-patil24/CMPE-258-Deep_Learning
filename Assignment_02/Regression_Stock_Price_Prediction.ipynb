```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, roc_curve, auc, precision_recall_curve
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import wandb
from wandb.integration.keras import WandbCallback  # Corrected import for WandbCallback

# Initialize Weights & Biases
wandb.init(project="stock-price-prediction", name="LSTM_Google_Stock")

# Load dataset
file_path = "/content/Google_Stock_Price_Train.csv"
df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Ensure df is a proper copy to avoid chained assignment issues
df = df.copy()

# Ensure numeric columns are properly formatted
df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')

# Handle NaN values in the DataFrame before scaling
df.ffill(inplace=True)  # Forward fill
df.bfill(inplace=True)  # Backward fill

# Check if the Volume column is entirely NaN and handle accordingly
print(df['Volume'].describe())
if df['Volume'].isnull().all():
    df.loc[:, 'Volume'] = 0  # If the entire column is NaN, replace with 0
else:
    df.loc[:, 'Volume'] = df['Volume'].fillna(df['Volume'].median())  # Normal median replacement

# Verify no NaN values remain
print("NaN count after Volume fill:", df['Volume'].isnull().sum())

# Ensure no missing values remain before proceeding
if df.isnull().sum().sum() > 0:
    print("Final NaN Check:\n", df.isnull().sum())
    raise ValueError("Data still contains NaN values after all filling attempts.")

# Selecting 'Open' as input and 'Close' as target
scaler = MinMaxScaler(feature_range=(0,1))
data_scaled = scaler.fit_transform(df[['Open', 'Close']])

# Create sequences for LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, 0])  # Open prices as features
        y.append(data[i+seq_length, 1])  # Close price as target
    return np.array(X), np.array(y)

seq_length = 50
X, y = create_sequences(data_scaled, seq_length)
X = X.reshape((X.shape[0], X.shape[1], 1))  # LSTM expects 3D input

# Splitting into training and test set
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Define LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(seq_length, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

model.save('my_model.keras')  # Save model in .keras format

history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[WandbCallback(log_graph=False)],
)

# Predictions
y_pred = model.predict(X_test)

y_pred = np.nan_to_num(y_pred)  # Handle NaN in predictions

# Reverse scaling
y_test_actual = scaler.inverse_transform(np.hstack((np.zeros((len(y_test), 1)), y_test.reshape(-1, 1))))[:, 1]
y_pred_actual = scaler.inverse_transform(np.hstack((np.zeros((len(y_pred), 1)), y_pred.reshape(-1, 1))))[:, 1]

# Compute evaluation metrics
mae = mean_absolute_error(y_test_actual, y_pred_actual)
mse = mean_squared_error(y_test_actual, y_pred_actual)
r2 = r2_score(y_test_actual, y_pred_actual)
wandb.log({"MAE": mae, "MSE": mse, "R2": r2})

# Plot results
plt.figure(figsize=(12,6))
plt.plot(df.index[-len(y_test):], y_test_actual, label='Actual Close Price')
plt.plot(df.index[-len(y_pred):], y_pred_actual, label='Predicted Close Price')
plt.legend()
plt.title("Stock Price Prediction using LSTM")
plt.show()

# ROC Curve (for classification tasks, adapted for regression error analysis)
fpr, tpr, _ = roc_curve(y_test_actual > np.mean(y_test_actual), y_pred_actual)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.title("ROC Curve for Stock Price Prediction")
plt.show()
wandb.log({"ROC AUC": roc_auc})

# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test_actual > np.mean(y_test_actual), y_pred_actual)
plt.figure()
plt.plot(recall, precision, color='green', label='Precision-Recall Curve')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.title("Precision-Recall Curve for Stock Price Prediction")
plt.show()
wandb.log({"Precision-Recall Curve": wandb.Image(plt)})

wandb.finish()
```
